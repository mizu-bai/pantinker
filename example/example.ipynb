```python
import pantinker
import scienceplots
from matplotlib import pyplot as plt
```

```python
df_npt_eq = pantinker.log_to_df("npt_eq.log")
df_npt_eq
```

```python
%matplotlib inline

with plt.style.context(["science", "no-latex"]):
    fig = plt.figure(dpi=150)
    ax = fig.add_subplot(111)
    ax.plot(df_npt_eq["Time"], df_npt_eq["Total Energy"])
    ax.set_xlabel("Time (ps)")
    ax.set_ylabel("Totol Energy (kJ/mol)")
```
